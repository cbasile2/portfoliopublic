```python
# import modules
from datetime import datetime
import yfinance as yf
import matplotlib.pyplot as plt
import copy
import pandas as pd
from tiingo import TiingoClient
from scipy.stats import ks_2samp
from dateutil import parser
import seaborn as sns
import pickle
import os

os.chdir("/Users/christianbasile/Desktop")

class Stock:
    def __init__(self, name, startdate, enddate):
        self.name = name
        self.startdate = startdate
        self.enddate = enddate

    def api_1(self):
        
        try:
            infile2 = open('api1_{0}.pkl'.format(self.name),'rb')
            table2 = pickle.load(infile2)
            infile2.close()
           
            
        except:
            print("Stock  name is " + self.name)
            config = {}
            # To reuse the same HTTP Session across API calls (and have better performance), include a session key.
            config['session'] = True
            # If you don't have your API key as an environment variable,
            # pass it in via a configuration dictionary.
            config['api_key'] = "9bdf418d9ef0e577ba126278e77230ca356d644e"
            # Initialize
            client = TiingoClient(config)

            # Get historical GOOGL prices from August 2017 as JSON, sampled daily
            historical_prices = client.get_ticker_price(self.name,
                                                        fmt='json',
                                                        startDate=self.startdate,
                                                        endDate=self.enddate,
                                                        frequency='daily')

            table2=pd.DataFrame(historical_prices)
            f = open("api1_{0}.pkl".format(self.name),"wb")
            pickle.dump(table2,f)
            f.close()
            print("saved")

        return table2

    def api_2(self):
        print("Stock  name is " + self.name)
        try:
            infile2 = open('api2_{0}.pkl'.format(self.name),'rb')
            data2 = pickle.load(infile2)
            infile2.close()
               
        except:
            
            data2 = pdr.get_data_yahoo(self.name, self.startdate, self.enddate)
            f = open("api2_{0}.pkl".format(self.name),"wb")
            pickle.dump(data2,f)
            f.close()
            print("saved")
        
        return data2

    def samedistribution(self):
        
        data1=self.api_1()
        data2=self.api_2()
        test = ks_2samp(data1['adjClose'],data2['Adj Close'])
        n1= len(data1)
        n2= len(data2)
        c =1.63 #0.01 cut off value
        Da= c*np.sqrt((n1 + n2)/(n1*n2))
        if test.statistic< Da:
            result='samedistribution'
            
        else:
             result='differentdistribution'
                
        x1=data1['adjClose']
        x2=data2['Adj Close']
        plt.hist(x1)
        plt.figtext(0.60,0.5, x1.describe().to_string())
        plt.xlabel("Price")
        plt.ylabel("Frequency")
        plt.show() 

        plt.hist(x2)
        plt.xlabel("Price")
        plt.ylabel("Frequency")
        #plt.figtext(0.60,0.5, x2.describe().loc[['mean','std']].to_string())
        plt.figtext(0.60,0.5, x2.describe().to_string())
        plt.show() 

            
        return result
    
    def addpercentdaily(self):
   
        try:
            infile2 = open('dailypercent{0}.pkl'.format(self.name),'rb')
            table2 = pickle.load(infile2)
            infile2.close()
          
        except:
            table2=self.api_1()
            #result= self.samedistribution()
            table2['month']=0
            table2['year']=0
            table2['day']=0
            table2['date'] = table2.apply(lambda row : parser.parse(row['date']), axis=1)
            table2['month'] = table2.apply(lambda row : row['date'].month, axis=1)
            table2['year'] = table2.apply(lambda row : row['date'].year, axis=1)
            table2['day'] = table2.apply(lambda row : row['date'].day, axis=1)
            table2['pDaily']=0
            for i in range(len(table2)-1):
                if i==0:
                    continue
                else:
                    table2['pDaily'].iloc[i] = ((table2['adjClose'].iloc[i]/table2['adjClose'].iloc[i-1]) -1)*100

   
            f = open('dailypercent{0}.pkl'.format(self.name),"wb")
            pickle.dump(table2,f)
            f.close()
            print("saved")
        
        return table2
    
    
    def bestmonths(self):
        
        tablef= self.addpercentdaily()
        describePrice=tablef['adjClose'].describe()
        describepDaily=tablef['pDaily'].describe()
        maskbest_pDaily=tablef['pDaily']> describepDaily['75%']
        maskworst_pDaily= tablef['pDaily'] < describepDaily['25%']
        maskbest_Price=tablef['adjClose']> describePrice['75%']
        maskbest= maskbest_pDaily*maskbest_Price
        tablef[maskbest].sort_values(by=['adjClose'], inplace=True)

        return tablef
       
        
    def worstmonths(self):
    
        tablef= self.addpercentdaily()
        describePrice=tablef['adjClose'].describe()
        describepDaily=tablef['pDaily'].describe()
        maskworst_pDaily= tablef['pDaily'] < describepDaily['25%']
        maskworst_Price= tablef['adjClose'] < describePrice['25%']
        maskworst= maskworst_pDaily*maskworst_Price
        tablef[maskbest].sort_values(by=['adjClose'], inplace=True)
        
        return tablef
    
    def groupyearmonth(self):
        tablef=self.addpercentdaily()
        groupby_yearmonth=tablef.groupby(by=["year","month"], dropna=False).last()
        groupby_yearmonth['monthreturn']=0
        startdateparser= parser.parse(self.startdate)
        enddateparser= parser.parse(self.enddate)
        years= range(startdateparser.year,enddateparser.year+1)
        months= range(startdateparser.month,enddateparser.month)
        for year in years:
            for month in months:
                if year==startdateparser.year and month==1:
                    continue

                else:

                    dap=(groupby_yearmonth['adjClose'].loc[year, month+1]/groupby_yearmonth['adjClose'].loc[year, month]-1)*100
                    groupby_yearmonth['monthreturn'].loc[year,month] = dap

        finale= groupby_yearmonth[['date','monthreturn']]
  
        return finale


    
    
    def groupyearmonthday(self):
    
        try:
            
            infile2 = open('groupyearmonthday_1.pkl','rb')
            ff = pickle.load(infile2)
            infile2.close()
            print("tried")
            
        except:

            tablef= self.addpercentdaily()
            print("load")
            tabletemp= tablef[['day','month','year','date','pDaily']]
            ff=tabletemp.groupby(['year','month','day']).first()
            f = open('groupyearmonthday_1.pkl',"wb")
            pickle.dump(ff,f)
            f.close()
            print("saved")

        return ff
    
    
    
    
    def outlier(self,target,year):
        tableff= self.addpercentdaily()
        
        tabletemp= tableff[['day','month','year','date','pDaily']]
        tablef=tabletemp.groupby(['year','month','day']).first()
        tablef= tablef.loc[year, :,:]
       
        tablef['z_score']=(tablef[target] - tablef[target].mean())/tablef[target].std()
        tablef_z= tablef[(tablef['z_score']>-3) & (tablef['z_score']<3)]
        describaa=tablef[target].describe()
        percent25= describaa['25%']
        percent75=describaa['75%']
        
        IQR= percent75-percent25
        lower_limit = percent25 - 1.5*IQR
        upper_limit = percent75 + 1.5*IQR
        
        tablef_iqr = tablef[(tablef[target]>lower_limit) & (tablef[target]<upper_limit)]
        
        """
        
        indices_z=[]
        indices_iqr=[]
        
        indices= list(tablef.index)
        indices_z= list(tablef_z.index)
        indices_iqr= list(tablef_iqr.index)
        
        for index in indices:
            if index not in indices_z:
                indices_z.append(index)
            if index not in indices_iqr:
                indices_iqr.append(index)
                
        removed_z=list(tablef.iloc[indices_z].index)
        removed_iqr= list(tablef.iloc[indices_iqr].index)
        """
    
        sns.boxplot(tablef[target]).set(title='for {} for year {} for variable {} '.format(self.name, year,target))
        
        dicresult= {
            'z': {'table':tablef_z},'iqr': {'table':tablef_iqr}
        }
        
        return dicresult
 
    
    def plot_price(self):
        # Display
        plt.figure(figsize = (20,10))
        plt.title('Opening Prices from {} to {} for {}'.format(self.startdate, self.enddate,self.name))
        table2= self.api_1()
        plt.plot(table2['date'],table2['adjClose'],color='blue')
        plt.savefig('plot_price.png', dpi=300, bbox_inches='tight')
        plt.show()
        
    def plot_percentdailyprice(self):
        # Display
        plt.figure(figsize = (20,10))
        plt.title('Opening Prices from {} to {} for {}'.format(self.startdate, self.enddate,self.name))
        table2= self.addpercentdaily()
        #plt.plot(data2.index,data2['Close'],color='red')
        plt.plot(table2['date'],table2['pDaily'],color='blue')
        plt.ylabel('%dailychange')
        plt.savefig('percentdaily.png', dpi=300, bbox_inches='tight')
        plt.show()
       
        
        
startdate = "2008-01-01"
enddate = "2021-12-31"        
name='GOOGL'
#Stock(name, startdate, enddate).groupyearmonth()

class Aggregate:
   
    def __init__(self, stocknames, startdate, enddate):
        self.stocknames = stocknames
        self.startdate = startdate
        self.enddate = enddate
        
    def aggregate_groupmonth(self):
        stocknames=self.stocknames
        stockname_1= stocknames[0]
        namef= 'monthreturn_{}'.format(stockname_1)
        Stockf= Stock(stockname_1, startdate, enddate)
        final= Stockf.groupyearmonth()       
        final.rename(columns={"monthreturn": namef},inplace=True)
        namet=[]
        namet.append(namef)
        for stockname in stocknames[1::]:
            name= 'monthreturn_{}'.format(stockname)
            namet.append(name)
            Stockf2= Stock(stockname, startdate, enddate)
            temp= Stockf2.groupyearmonth()
            final[name]=0
            final[name]=temp['monthreturn']
            
        return final[namet]
    
    
    def aggregate_groupmonthday(self):
        
        stocknames=self.stocknames
        stockname_1= stocknames[0]
        name='pDaily_{}'.format(stockname_1)
        Stockf= Stock(stockname_1, startdate, enddate)
        tabletemp= Stockf.groupyearmonthday()
  
        tabletemp.rename(columns={"pDaily": name},inplace=True)
        namet=[]
        namet.append(name)
    
        
        for stockname in stocknames[1::]:
            
            namei='pDaily_{}'.format(stockname)
            namet.append(namei)
            tabletemp[namei]=0
            Stockf= Stock(stockname ,startdate, enddate)
            final= Stockf.groupyearmonthday()
            tabletemp[namei]=final['pDaily']
            
        ff=tabletemp.groupby(['year','month','day']).first()
        return ff[namet]
    
    def corr_months(self,year):
        final=self.aggregate_groupmonth()
        finale=final.loc[year, :].corr(method='pearson')
        return finale
        
    def corr_day(self,year,month):
        final=self.aggregate_groupmonthday()
        finale=final.loc[year, month,:].corr(method='pearson')
        return finale
    
    def plotmonth(self, year):

        #Month Section
        titlep= 'PlotMonth'
        stocknames= self.stocknames
        monthvaluesdic= {}
        for stockname in stocknames:
            monthvaluesdic[stockname]=[]
            
        labelssample= [i for i in range(1,13)]
        labels= stocknames
        
        update_maxvale= 0
        update_minvale=-10
        for stockname in stocknames:
            for month in range(1,13):
                name='monthreturn_{}'.format(stockname)
                table=self.aggregate_groupmonth()
        
                maxvale=max(table.loc[year,:][name])
                minvale= min(table.loc[year,:][name])
                
                if maxvale>update_maxvale:
                    update_maxvale=maxvale
                if minvale < update_minvale:
                    update_minvale= minvale
                    
                    
                monthvaluesdic[stockname].append(table.loc[year,month][name])

        barWidth = 0.1
        fig = plt.figure()
        fig.set_figheight(10)
        fig.set_figwidth(20)
        fig.subplots_adjust(hspace=0, wspace=0)
        ax = fig.add_subplot(1, 2, 1)
        r1 = np.arange(len(monthvaluesdic[stocknames[0]]))
        r2 = [x + barWidth for x in r1]
        r3 = [x + barWidth for x in r2] 
        
        colors= ['red','blue','green']
        rs=[r1,r2,r3]
        for ix,stockname in enumerate(stocknames):
    
            ax.bar(rs[ix], monthvaluesdic[stockname],  label=stockname, width=barWidth, color=colors[ix])
     
        tochange_title = 'Month Return Comparisons for {}'.format(year)
        ax.set_xticks(np.arange(len(labelssample)))
        ax.set_xticklabels(labelssample, rotation='horizontal', fontsize=14, fontweight='bold')
        ax.set_xlabel('Months', fontsize=16)
        ax.set_title(tochange_title, fontsize=14, fontweight='bold')
        ax.legend(labels,bbox_to_anchor=(1.10,-0.08), fontsize=16)
        ax.set_ylim(update_minvale, update_maxvale)
        ax.set_ylabel('%MonthReturn')
        ax.legend()
        fig.suptitle('Month Metrics', fontsize=18)
        fig.legend()
        to_savepng= 'plotmonth.png'
        fig.savefig(to_savepng, dpi=300, bbox_inches='tight')
        fig.show()
        
    def plotmonthday(self, year,month):
        #Month Section
        titlep= 'PlotMonthDay'
        stocknames= self.stocknames
        table=self.aggregate_groupmonthday()
        daylabels=list(table.loc[year,month].index)
        #Day Section
        dayvaluesdic= {}
        for stockname in stocknames:
            dayvaluesdic[stockname]=[]
            
        labels= stocknames
        for stockname in stocknames:
            for day in daylabels:
                if stockname==stocknames[0]:
                    try:
                        name='pDaily_{}'.format(stockname)
                    
                    except:
                        name='pDaily'
                        
                else:
                    name='pDaily_{}'.format(stockname)

                daylabels.append(day)
                dayvaluesdic[stockname].append(table[name].loc[year,month,day])


        barWidth = 0.1
        fig = plt.figure()
        fig.set_figheight(10)
        fig.set_figwidth(20)
        fig.subplots_adjust(hspace=0, wspace=0)
        ax = fig.add_subplot(1, 2, 1)
        
        batches= np.array_split(np.array(dayvaluesdic[stockname]),4)
        
        
        for index,batch in enumerate(batches):
            print(index)
            

            r1 = np.arange(len(dayvaluesdic[stocknames[0]]))
            r2 = [x + barWidth for x in r1]
            r3 = [x + barWidth for x in r2]
            colors= ['red','blue','green']   
            rs=[r1,r2,r3]

            for ix,stockname in enumerate(stocknames):
                print(stockname)

                ax.bar(rs[ix], dayvaluesdic[stockname],  label=stockname, width=barWidth, color=colors[ix])

            labels= stocknames
            tochange_title = 'Day Return Comparisons for {0} and {1}'.format(year, month)
            ax.set_xticks(np.arange(len(daylabels)))
            ax.set_xticklabels(daylabels, rotation='vertical', fontsize=7, fontweight='bold')
            ax.set_xlabel('Days', fontsize=16)
            ax.set_title(tochange_title, fontsize=14, fontweight='bold')
            ax.legend(labels,bbox_to_anchor=(1.10,-0.08), fontsize=16)
            ax.set_ylim(-50, 50)
            ax.set_ylabel('%DailyReturn')
            ax.legend()
            fig.suptitle('Day Metrics', fontsize=18)
            fig.legend()
            to_savepng= 'plotmonth{}.png'.format(index)
            fig.savefig(to_savepng, dpi=300, bbox_inches='tight')
            fig.show()

print("Done")

#final.columns= ['date', 'monthreturn_GOOGL', 'monthreturn_AMZN', 'monthreturn_WFC']
stocknames= ['GOOGL', 'AMZN','WFC']
Stock(stocknames[0], startdate, enddate).addpercentdaily()





#Stock(stocknames[0], startdate, enddate).groupyearmonth() 
#Aggregate(stocknames, startdate, enddate).aggregate_groupmonth() 
#Aggregate(stocknames, startdate, enddate).aggregate_groupmonthday()
#Aggregate(stocknames, startdate, enddate).plotmonthday(2008,1)
#Aggregate(stocknames, startdate, enddate).plotmonth(2008)
#Aggregate(stocknames, startdate, enddate).aggregate_groupmonth()
#tablef=self.addpercentdaily()
#for stock in stocknames:
#    Stockf2= Stock(stock, startdate, enddate).plot_price()
#    Stockf2= Stock(stock, startdate, enddate).plot_percentdailyprice()
    
#Aggregate(stocknames, startdate, enddate).plotmonth(2021)
ff=Stock(stocknames[0], startdate, enddate).addpercentdaily()


#Stock(stocknames[0], startdate, enddate).groupyearmonth() 
#addpercentdaily(self)
#Aggregate(stocknames, startdate, enddate).plotmonthday(2008,1)
targets= ['adjClose','pDaily']


pp=Stock(stocknames[2], startdate, enddate).outlier(targets[1],year)
#pp=Stock(stocknames[1], startdate, enddate).samedistribution()


year=2021
month=10
#Aggregate(stocknames, startdate, enddate).corr_months(year)
#month=10



ff=Aggregate(stocknames, startdate, enddate).corr_day(year,month)

final=Aggregate(stocknames, startdate, enddate).aggregate_groupmonthday()
finale=final.loc[year, month,:].corr(method='pearson')

final

#Aggregate(stocknames, startdate, enddate).aggregate_groupmonthday()

#groupyearmonth, 
#groupyearmonthday, outlier(self,target,year), 
#plot_price, plot_percentdailyprice

#ff=Stock(stocknames[2], startdate, enddate).plot_percentdailyprice()

#Aggregate(stocknames, startdate, enddate).plotmonth(2022)

ff=Stock(stocknames[0], startdate, enddate).addpercentdaily()
ff

groupm= Stock(stocknames[0], startdate, enddate).groupyearmonth()

groupm


groupm= Stock(stocknames[0], startdate, enddate).groupyearmonth()
groupymd=Stock(stocknames[0], startdate, enddate).groupyearmonthday()
groupymd.loc[2008,1,:]
```
