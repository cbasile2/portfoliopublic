```python
import cv2
import numpy as np

import cv2
import numpy as np

# Create a black image
img = np.zeros((500, 500), dtype=np.uint8)

# Draw a white circle in the middle
cv2.circle(img, (250, 250), 100, (255), -1)

# Apply Gaussian blur to the image0
blur = cv2.GaussianBlur(img, (5, 5), 0)

# Perform Canny edge detection
edges = cv2.Canny(blur, 50, 150)

# Perform Hough Circle Transformation
circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, dp=1, minDist=100, param1=50, param2=30, minRadius=0, maxRadius=0)

if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        cv2.circle(img, (i[0], i[1]), i[2], (255), 2)

# Display the final image with circles
cv2.imshow("Image with circles", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Create a black image
img = np.zeros((500, 500), dtype=np.uint8)

# Draw a white line in the middle
cv2.line(img, (100, 250), (400, 250), (255), 5)

# Apply Gaussian blur to the image
blur = cv2.GaussianBlur(img, (5, 5), 0)

# Perform Canny edge detection
edges = cv2.Canny(blur, 50, 150)

# Display the edge detected image
cv2.imshow("Edge detected image", edges)
cv2.waitKey(0)

# Perform Hough Line Transformation
lines = cv2.HoughLines(edges, 1, np.pi/180, 200)

if lines is not None:
    for line in lines:
        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))

        cv2.line(img,(x1,y1),(x2,y2),(255),2)

# Display the final image with lines
cv2.imshow("Image with lines", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
